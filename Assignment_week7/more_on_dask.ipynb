```python
import pandas as pd
import dask.dataframe as dd
import time
from dask.distributed import Client
client = Client(n_workers=4) #cuz four cores/threads
client
```

```python
df = pd.read_csv("test/groupby-N_1000000_K_100_file_0.csv")
ddf = dd.read_csv("test/groupby-N_1000000_K_100_file_0.csv")
```

```python
def perform_test(df):
    return df.groupby("id1")["v1"].sum()

t = time.time()
dask_test = perform_test(ddf).compute()
d_t = time.time() - t
t = time.time()
pd_test = perform_test(df)
pd_t = time.time() - t
```

```python
print(f'Pandas test took: {pd_t}\nDask test took: {d_t}')
print(f'dask_test = {isinstance(dask_test,dd.DataFrame)}\npd_test = {isinstance(pd_test, pd.DataFrame)}')
```
